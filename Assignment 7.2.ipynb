## <u>Qus- 1</u>:- What are the three stages to build the hypotheses or model in machine learning?

**Ans-1 :-** Fallowing are the three stages to build the model in machine learning :-<br>

   &emsp; &emsp; 1- Data Preprocessing :- when we get a set of data to build the machine learning model, it requires different pre-processing action to clean the data like formatting, cleaning and sampling.<br>
   
   &emsp; &emsp; 2- Build and Testing the model :- when our data gets cleaned up we apply a fraction of that data on an suitable machine learing algorithm. algorithm will train itself by that data and then builds a model. after model formation we apply unseen data to the model to test the model.
   
   &emsp; &emsp; 3- Applying the model :- After building and testing the model we apply that model in the real world problems.

## <u>Qus- 3</u>:-What is Training set and Test set?

**Ans-3 :-** when we get a data set we divide them into Training and Test set.<br>
    &emsp; &emsp; **Training set :-** it is a dataset we use to train a model. we apply this data set onto an machine learning algorithm. if the training set is labeled correctly then the model should be able to learn the relationship between the data value(input) and their corresponding label(output).
    
   &emsp; &emsp; **Test set :-** it is uses to test the accuracy of prediction generated by the model. it is the unseen dataset to the model. so we apply it on the model and see how well it predicted all values.

## <u>Qus-4</u> :- What is the general principle of an ensemble method and what is bagging and
   ## boosting in ensemble method?

**Ans-4 :-**  <br>
**Ensemble Learning method** :- It means grouping multiple week learning models to form a strong model so that we can obtain better prediction.

**Begging :-**  It is uses in Low Bias - High Variance problem. this problem occur when the model overfits like in Decision tree.<br>
* In this case various models are built in parallel and each model gets trained on randomly selected samples.<br>
* then the various models vote to give the final prediction.<br>
* predictions will be averaged to get the final prediction(in case of regression) and in case of classification the final prediction will be the mode of the predicted ans.<br>

**Boosting :-** It is uses in Low variance - High Bias problem. this problem occures when model underfits.<br>
   1. In this method we first sampled the input data to generate a set of training data.
   2. then we run an algorithm on this training data to get a trained model. 
   3. then we take all our training data to test the model and we are gonna discover that some of the points are not well predicted.
   4. now we'll gonna go to build the second bag of sampled data. In this also the data will be randomly choosen, but now each data point is weighted according to error found in last model. so these values are more likely to get picked in this bag than any other else.
   5. now we'll build a model for this sample set also then we'll test it.<br>
       here the testing will be performed on both of the model and the result will be mode of the both result( in case of classification) in case of regression result will be mean of the both result.<br>
   6. now again we'll find some values which are not predicted well. so we'll build one more bag and one more model and this process will continue.

## <u>Qus-5</u>:- How can you avoid overfitting?

**Ans-5 :-** when we train the model on training set to such a level that it starts predicting noise or outlier present in training data correctly. then we say that model gets overfit. because in this case model will show 100% accuracy for training set but its accuracy will be low on test data.<br>
there are multiple methods by which we can avoid overfitting -<br>
    1- Ensamble method :- it is the best method to avoid overfitting and increasing accuracy.<br>
    2- Remove features :- its example is Pruning, which we do in case of decision tree model.<br>
    3- Cross validation
